## Applications to Memory Allocation

### Example: Pre-allocating Memory

Suppose we wish to create a large list of 1,000,000 numbers. The simplest way is to just run a loop, adding one element at a time to the vector. (Note that the runtime will vary depending on the hardware where you run the code.)

```python
import time

number = 1000000

# Check the current time
startTime = time.time()

# Create an empty list
list = []

# Add items to the list one by one
for counter in range(number):
    list.append(counter)

# Display the run time
print(time.time() - startTime)
```

The code below can pre-allocate an array with 1,000,000 and fill it in with 1, 2, ..., 1,000,000. (Note that the runtime will vary depending on the hardware where you run the code.)

```python
import time
number = 1000000

# Check the current time
startTime = time.time()

# Create a list of 1000000 zeros
list = [None]*number

# Add items to the list one by one
for counter in range(number):
    list[counter] = counter

# Display the run time
print(time.time() - startTime)
```

## Efficacy of Brute Force Algorithms

### Example: Caesar Cipher

A brute force check of how the text would be decoded for each possible Caesar cipher.

```python
# Intercepted message
codedMessage = 'nzohfu gur rarzl ng avtug'

# We will shift by 0, shift by 1, shift by 2, ... and print the results
for counter in range(26):
    # Start with no guess
    guessedMessage = ''

    # Loop through each letter in the coded message
    for x in codedMessage:

        # If x is not a space
        if x != ' ':

            # Shift the letter forward by counter
            if ord(x)+counter <= 122:
                x = chr(ord(x)+counter)

            # Subtract 26 if we go beyond z
            else:
                x = chr(ord(x)+counter-26)

        # Build a guess for the message one letter at a time
        guessedMessage = guessedMessage + x

    # Print the counter (the shift) and the message
    print(counter, guessedMessage)
```

Clearly, the length of the shift is 13 letters as only this one produces an intelligible message.

### Random Substitution Cipher

This code implements a simple random substitution cipher. Each letter in the alphabet is randomly mapped to another unique letter. The original message is then encrypted using this mapping. Non-alphabetic characters (like spaces) are kept as they are.

```python
import random
import string

def random_encrypt(message):
    # Create a list of all lowercase letters
    alphabet = list(string.ascii_lowercase)
    # Create a shuffled list of letters to serve as the substitution map
    shuffled_alphabet = random.sample(alphabet, len(alphabet))

    # Create the substitution dictionary
    # e.g., {'a': 'q', 'b': 'z', ...}
    substitution_map = dict(zip(alphabet, shuffled_alphabet))

    encrypted_message = []
    for char in message:
        if 'a' <= char <= 'z':
            # If it's a lowercase letter, substitute it
            encrypted_message.append(substitution_map[char])
        elif 'A' <= char <= 'Z':
            # If it's an uppercase letter, convert to lowercase for lookup,
            # then convert the substituted letter back to uppercase
            lower_char = char.lower()
            substituted_lower = substitution_map[lower_char]
            encrypted_message.append(substituted_lower.upper())
        else:
            # If it's not a letter, keep it as is (e.g., spaces, punctuation)
            encrypted_message.append(char)

    return ''.join(encrypted_message)

# The message to encrypt
message_to_encrypt = "next wednesday is math exam"

# Encrypt the message
encrypted_text = random_encrypt(message_to_encrypt)

print(f"Original message: {message_to_encrypt}")
print(f"Encrypted message: {encrypted_text}")
```

### Updated Random Substitution Cipher with Decryption

To allow for decryption, the `random_encrypt` function now returns the `substitution_map`. A new `random_decrypt` function is provided to reverse the process using this map.

```python
import random
import string

def random_encrypt_with_key(message):
    # Create a list of all lowercase letters
    alphabet = list(string.ascii_lowercase)
    # Create a shuffled list of letters to serve as the substitution map
    shuffled_alphabet = random.sample(alphabet, len(alphabet))

    # Create the substitution dictionary
    substitution_map = dict(zip(alphabet, shuffled_alphabet))

    encrypted_message_chars = []
    for char in message:
        if 'a' <= char <= 'z':
            encrypted_message_chars.append(substitution_map[char])
        elif 'A' <= char <= 'Z':
            lower_char = char.lower()
            substituted_lower = substitution_map[lower_char]
            encrypted_message_chars.append(substituted_lower.upper())
        else:
            encrypted_message_chars.append(char)

    return ''.join(encrypted_message_chars), substitution_map

def random_decrypt_with_key(encrypted_message, substitution_map):
    # Create a reverse substitution dictionary for decryption
    # e.g., {'q': 'a', 'z': 'b', ...}
    reverse_substitution_map = {v: k for k, v in substitution_map.items()}

    decrypted_message_chars = []
    for char in encrypted_message:
        if 'a' <= char <= 'z':
            decrypted_message_chars.append(reverse_substitution_map[char])
        elif 'A' <= char <= 'Z':
            lower_char = char.lower()
            decrypted_message_chars.append(reverse_substitution_map[lower_char].upper())
        else:
            decrypted_message_chars.append(char)

    return ''.join(decrypted_message_chars)

# The message to encrypt
message_to_encrypt = "next wednesday is math exam"

# Encrypt the message and get the key
encrypted_text, key = random_encrypt_with_key(message_to_encrypt)

print(f"Original message: {message_to_encrypt}")
print(f"Encrypted message: {encrypted_text}")

# Decode the message using the obtained key
decrypted_text = random_decrypt_with_key(encrypted_text, key)
print(f"Decrypted message: {decrypted_text}")
```
